```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
# aca establezco en que pais e idioma voy a buscar toda la info
TrendReq(hl='es-AR')
```

```python
# aca pongo la palabra para cual voy a solicitar toda la info
kw_list = ["articulos hogar"]
pytrend.build_payload(kw_list,geo='AR',timeframe='today 1-m')#,cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python

# Interest by Region
df = pytrend.interest_by_region(resolution = 'Argentina')
df = df.sort_values('articulos hogar', ascending=False)
df.head(10)
```

```python
df.reset_index().plot(x='geoName', y='articulos hogar', figsize=(120, 10), kind ='bar')
```
